```
from bs4 import BeautifulSoup
```

```
import requests
```

```
url= 'https://en.wikipedia.org/wiki/Nicolaus_Copernicus'
```

```
# we are sending request to the url and taking the responce
responce=requests.get(url)
responce
```

```
responce.text
```

```
# parsing library html.parser and second is lxml
Soup=BeautifulSoup(responce.text,"lxml")
```

```
Soup
```

```
for Alink in Soup.find_all('a'):
  #print(Alink)
  print(Alink.get('href'))
```

```
for ptag in Soup.find_all('p'):
  #print(ptag)
  # now second loop for btag
  for btag in ptag.find_all('b'):
    print(btag.text)
```
